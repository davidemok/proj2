## Part 3: Execute Data Cleaning Plan Using DEFINE / CLEAN / TEST Approach

### Step 1: Understand DEFINE / CLEAN / TEST Approach

Refer to project instructions for an explanation of this step. You can copy and past the template in the cell below to use in each step of Part 3 as you are treating the issues you found in the data. 

```python
### DEFINE ###

#drop unnamed column
### CLEAN ### 
house_df.drop(columns=['Unnamed: 0'], inplace=True)

### TEST ###
house_df.head()
```

```python
### DEFINE ###
#rename waterfront
#drop unnamed column
### CLEAN ### 
house_df.rename(columns={'Waterfront?': 'Waterfront'}, inplace=True)

### TEST ###
house_df.head()
```

```python
### DEFINE ###

# replace empty spaces with _
### CLEAN ### 
house_df.columns = house_df.columns.str.replace(" ", "_")

### TEST ###
house_df.head()
```

```python
### DEFINE ###
#zipcode and ID turned to string

### CLEAN ### 
house_df = house_df.astype({"Zipcode": str, "ID": str})

### TEST ###
house_df.head()
```

### Step 2: Treat Names / Casing / Spelling Issues

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###
#lowercase

### CLEAN ### 
house_df.columns = house_df.columns.str.lower()

### TEST ###
house_df.head()
```

```python
sales_df.columns = sales_df.columns.str.lower()
sales_df.head()
```

```python
sales_df.drop(columns = ['unnamed: 0'], inplace = True)
sales_df.head()
```

```python
sales_df.price = sales_df.price.str.replace("$", "")
sales_df.head()
```

### Step 3: Treat Missing Values

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###
#change empty spaces in string columns to unknown or something that makes sense

### CLEAN ### 
house_df.house_condition = house_df.house_condition.fillna('unknown')
house_df.waterfront = house_df.waterfront.fillna('unknown')
house_df.year_renovated = house_df.year_renovated.fillna('not renovated')
house_df.lat = house_df.lat.fillna('unknown')
house_df.long = house_df.long.fillna('unknown')

### TEST ###
house_df.isna().sum()
```

```python
house_df.view = house_df.view.fillna('so bad it was left blank')
house_df.view.isnull().sum()
```

```python
# for the rest of the numerical columns i will just add the mean to the empty spaces so I don't wreck any plots later. 
#i suppose i could also add the mode or median (actually median is probably best since it ignores outliers but whatever haha)

house_df.bathrooms = house_df.bathrooms.fillna(house_df.bathrooms.mean())
house_df.sqft_lot = house_df.sqft_lot.fillna(house_df.sqft_lot.mean())
house_df.grade = house_df.grade.fillna(house_df.grade.mean())
house_df.sqft_above = house_df.sqft_above.fillna(house_df.sqft_above.mean())
house_df.year_built = house_df.year_built.fillna(house_df.year_built.mean())

house_df.isna().sum()
```

```python
sales_df.isna().sum()
```

```python
sales_df.price = sales_df.price.str.replace(",", "")
sales_df.price
```

```python
sales_df = sales_df.astype({"price": float})
```

```python
sales_df.price = sales_df.price.fillna(sales_df.price.mean())
```

```python
sales_df.price.isna().sum()
```
